1 Program To Print ASCII Value of a Character.

```python
c = 'a'
print("The ASCII value of '" + c + "' is", ord(c))
```

2 Print ASCII Value of All Characters.

```python
for i in range(256):
    print(f"ASCII value of {chr(i)} is {i}")
```

3 Find the Largest Among Three Numbers.

```python
a1 = input('Enter the first num:')

a2 = input('Enter the second num:')

a3 = input('Enter the third num:')

if int(a1)>int(a2) and int(a1)>int(a3):
     print('The largest is:',a1)
   
    
elif int(a2)>int(a3) and int(a2)>int(a1):
    print('The largest is:',a2)
    
else:
    print('The largest is:',int(a3))
    
```

5 Program To Check Leap Year Or Not.

```python
leapyr = int(input('Enter the year:'))

if (leapyr%4)==0:
    print('The entered year is a leap year:',leapyr)
    
else:
    print('This is not a leap year:',leapyr)
```
